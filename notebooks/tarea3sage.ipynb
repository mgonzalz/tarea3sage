```python
#Calcular todas las soluciones de la ecuación x^3 − x = 7*x^2 − 7.
x = var("x")
solve(x^3 - x == 7*x^2 - 7, x)
```

```python
#Calcular el conjunto solución de la inecuación |t − 7| ≥ 3.
t = var('t')
solve(abs(t - 7)>= 3, t)
```

```python
#Calcular todas las soluciones del sistema 2x+y=17, x−3y=−16.
x, y = var('x,y')
solve([2*x + y == 17, x - 3*y == -16], x, y)
```

```python
#Definiendo la función junto a su gráfica.
x = var("x")
f(x) = exp(x) - cos(x)
plot(f(x), x, [-pi/2, 0])
```

```python
#Usar find root() para calcular una solución de la ecuación exp(x) = cos(x) en el interval0 [−π/2, 0].
find_root(f(x),-pi/2, 0)
```

```python
#Cambia el comando anterior para que find_root calcule otra solución en el mismo intervalo.
find_root(f(x), -pi/2, -0.001)
```
